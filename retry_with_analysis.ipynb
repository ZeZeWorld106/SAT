```python
# win_sizecoherence_mean
win_size = 10
frame_num = len(records['frameidx'])
trial_num = round(frame_num/win_size)


coherences = []
for i in range(trial_num):
    #   (0-10, 10-20, 20-30...)
    win_enemy = records['enemydotnum'][win_size*i : win_size*(i+1)]
    win_allay = records['allaydotnum'][win_size*i : win_size*(i+1)]
    #  coherence
    coherence = np.round(np.mean(win_enemy)/(np.mean(win_enemy) + np.mean(win_allay)), 3)
    #  coherence
    coherences.append(coherence)

# for i in records['frameidx']:
#     if i % win_size == 0:
        
#         curr_mean_coherence_ratio = (np.array(records['enemydotnum'])/(np.array(np.array(records['enemydotnum']) + np.array(records['allaydotnum']))) / win_size
#         curr_coherence_ratio_correction = np.sort(np.round(curr_coherence_ratio * 100, 2), axis=0)
    
# coherence = curr_coherence_ratio_correction
# win_size
add_hit = []
minus_hit = []
no_resp = []
for i in range(trial_num):
    #  response
    win_response = records['response'][win_size*i : win_size*(i+1)]
    #  add response
    add_num = np.sum(np.array(win_response)==1)
    #  minus response
    minus_num = np.sum(np.array(win_response)==-1)
    # no response
    cur_no_resp = int(np.sum(win_response) == 0) 
    #  response
    add_hit.append(add_num)
    minus_hit.append(minus_num)
    no_resp.append(cur_no_resp)

# for i in records['frameidx']:
#   if i % win_size == 0:

#     curr_add_hit = (records['response'][i:i+win_size]).count(1)
#     add_hit.append(curr_add_hit)
    # curr_min_hit = (records['response'][i:i+win_size]).count(-1)
  
    # print(f"{i}{i+win_size}+:{curr_add_hit}, -:{curr_min_hit}")
# choices_to_add = np.sort(np.array(add_hit), axis=0)
# len(add_hit)
```

```python
win_idx = np.argsort(coherences)
coherences = np.sort(coherences)
add_hit = np.array(add_hit)[win_idx]
minus_hit = np.array(minus_hit)[win_idx]
no_resp = np.array(no_resp)[win_idx]
#  coherences ,  num_bins 
num_bins = 10
trial_num, bin_range = np.histogram(coherences, bins=num_bins)
add_resposne_rates = []
minus_response_rates = []
no_response_rates = []
mean_coherences = []
for i in range(num_bins):
    # bin
    bin_lower_bound, bin_upper_bound = bin_range[i : i+2]
    # bincoherence trial
    trial_within_bin = np.where(np.logical_and(coherences>=bin_lower_bound, coherences<bin_upper_bound)==1)[0]
    if i == num_bins - 1:
        trial_within_bin =  np.where(np.logical_and(coherences>=bin_lower_bound, coherences<=bin_upper_bound)==1)[0]
    # bin + 
    add_resposne_rate = np.mean(add_hit[trial_within_bin]>0)
    add_resposne_rates.append(add_resposne_rate)
    # bin - 
    minus_response_rate = np.mean(minus_hit[trial_within_bin]>0)
    minus_response_rates.append(minus_response_rate)
    # bin  
    no_response_rate = np.mean(no_resp[trial_within_bin]>0)
    no_response_rates.append(no_response_rate)
    # bin
    # mean_coherences.append(np.mean([bin_lower_bound, bin_upper_bound]))
    mean_coherence = np.mean(coherences[trial_within_bin])
    mean_coherences.append(mean_coherence)

# plt.plot(mean_coherences, add_resposne_rates, marker='o',markersize=10)
plt.plot(mean_coherences, np.array(minus_response_rates) + np.array(add_resposne_rates), marker='*',markersize=10)

# plt.plot(mean_coherences, minus_response_rates, marker='*',markersize=10)
plt.plot(mean_coherences, no_response_rates, marker='^',markersize=10)
# plt.hlines(0.5, 0,1)
plt.xlim([0,1])
```
